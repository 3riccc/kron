```python
l0 = []
l1 = []
l2 = []
```

```python
# roll
for line in lines:
    try:
        if 'train loss:' in line:
            l00 =  float(line.split('train loss:')[-1])
            l0.append(l00)
        if 'auc' in line:
            l11 =  float(line.split('auc:')[1])
            l1.append(l11)
    except:
        continue
```
