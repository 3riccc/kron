```python
import matplotlib.pyplot as plt
import numpy as np
add = '../log/kron36.txt'
f = open(add)
lines = f.readlines()
for line in lines:
    if 'start_time' in line:
        print(line)
    if 'end_time' in line:
        print(line)
    if 'base' in line:
        print(line)
    if 'test' in line:
        print(line)
    if 'gn_hid' in line:
        print(line)
    if 'hid mask' in line:
        print(line)
```

```python
l0 = []
l1 = []
l2 = []
```

```python
# roll
for line in lines:
    try:
        if 'train loss:' in line:
            l00 =  float(line.split('train loss:')[-1])
            l0.append(l00)
        if 'auc' in line:
            l11 =  float(line.split('auc:')[1])
            l1.append(l11)
        if 'train structural loss' in line:
            l22 = float(line.split('train structural loss:')[1])
            l2.append(l22)
    except:
        continue
```

```python
plt.figure(figsize=(8,4))
plt.subplot(121)
plt.plot(l0,color='red')
plt.xlabel('train epoch')
plt.yscale('log')
plt.title('train mse loss')
plt.subplot(122)
plt.plot(l1,color='blue')
plt.xlabel('train epoch')
plt.title('auc')
# plt.plot([500,500],[min(l1),max(l1)+0.02],':',color='lightgreen')
plt.show()
```
